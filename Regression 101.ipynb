```julia
using DataFrames
using CSV
using Plots
using Lathe
using GLM
using Statistics
using StatsPlots
using MLBase
```

```julia
df = DataFrame(
  Dependent = [5,6,2,4,5,23,4,5,5,5,4,3,2,43,2],
  Independent = [50,6,20,4,5,230,4,5,5,15,4,13,2,43,20]
)
```

```julia
fm = @formula(Dependent ~ Independent)
linearRegressor = lm(fm, df)
```

```julia
r2(linearRegressor)
```

```julia
cor(df.Independent,df.Dependent)
```

```julia
scatter(df.Independent,df.Dependent, title = "Dependent vs Independent", ylabel = "Dependent", xlabel = "Independent",legend = false)
```
